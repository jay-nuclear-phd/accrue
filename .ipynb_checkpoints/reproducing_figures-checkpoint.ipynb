```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random

file_path = r"C:\Users\js229662\OneDrive - The University of Texas at Austin\Library\VALID-master\benchmark_620.pkl"

with open(file_path, 'rb') as f:
    data = pickle.load(f)

m = data['meas_keff']    # measured keff
k = data['calc_keff']    # calculated keff
Cm = data['meas_cov']    # measured covariance
Ck = data['calc_cov']    # calculated covariance
names = data['names']
```

```python
total_indices = list(range(620))

n = 1500
random.seed(1)
res = np.zeros((n,3))
for i in range(n):
    selected_indices = random.sample(total_indices, 11)

#     print("Selected indices:", selected_indices)

    rk = k[selected_indices]
    rm = m[selected_indices]
    rCk = Ck[np.ix_(selected_indices, selected_indices)]
    rCm = Cm[np.ix_(selected_indices, selected_indices)]

    std_dev = np.diag(np.sqrt(np.diag(rCk)))
    Rc = np.linalg.inv(std_dev)@rCk@np.linalg.inv(std_dev)
    ck_all = Rc[:-1,-1]
    ck = np.mean(ck_all)

    Rd = np.linalg.inv(std_dev)@(rCk + rCm)@np.linalg.inv(std_dev)
    jk = np.sqrt(Rd[:-1,-1]@np.linalg.inv(Rd[:-1,:-1])@Rd[:-1,-1])

    pk = rk[-1] - rCk[-1,:-1]@np.linalg.inv(rCk[:-1,:-1] + rCm[:-1,:-1])@(rk[:-1] - rm[:-1])
    pCk = rCk[-1,-1] - rCk[-1,:-1]@np.linalg.inv(rCk[:-1,:-1] + rCm[:-1,:-1] + np.eye(10)*1e-10)@rCk[-1,:-1]
    pu = np.sqrt(pCk)

    score = (pk - rm[-1]) / pu

    ck, jk, score
    res[i,:] = np.array([ck, jk, score])
    
plt.figure(figsize=(8, 6))
plt.plot(res[:, 0], res[:, 1], '.')

plt.xlim([0, 1])
plt.ylim([0, 1])

plt.xticks(np.arange(0, 1.01, 0.1))
plt.yticks(np.arange(0, 1.01, 0.1))

plt.xlabel(r'average $c_k$', fontsize=12)
plt.ylabel(r'$j_k$', fontsize=12)

plt.grid()
plt.show()
```

```python
matching_seeds_only = []

for seed in range(1000):
    total_indices = list(range(620))
    random.seed(seed)
    selected_indices = random.sample(total_indices, 100)

    rk = k[selected_indices]
    rm = m[selected_indices]
    rCk = Ck[np.ix_(selected_indices, selected_indices)]
    rCm = Cm[np.ix_(selected_indices, selected_indices)]
    rCd = rCk + rCm
    rNames = [names[i] for i in selected_indices]

    std_dev = np.diag(np.sqrt(np.diag(rCd)))
    with np.errstate(divide='ignore', invalid='ignore'):
        inv_std = np.linalg.inv(std_dev + 1e-12 * np.eye(std_dev.shape[0]))  #  

    Rd = inv_std @ rCd @ inv_std

    row = Rd[0, :].copy()
    row[0] = np.inf
    closest_idx = np.argmin(np.abs(row - 0.7))

    if not (0.69 <= Rd[0, closest_idx] <= 0.71):
        continue

    remaining_indices = [i for i in range(100) if i not in (0, closest_idx)]
    random.shuffle(remaining_indices)
    new_order = [0, closest_idx] + remaining_indices

    rk = rk[new_order]
    rm = rm[new_order]
    rCk = rCk[np.ix_(new_order, new_order)]
    rCm = rCm[np.ix_(new_order, new_order)]
    rCd = rCk + rCm
    rNames = [rNames[i] for i in new_order]

    std_dev = np.diag(np.sqrt(np.diag(rCd)))
    inv_std = np.linalg.inv(std_dev + 1e-12 * np.eye(std_dev.shape[0]))
    Rd = inv_std @ rCd @ inv_std

    total_jk = np.sqrt(Rd[1:, 0] @ np.linalg.inv(Rd[1:, 1:]) @ Rd[1:, 0])
    if total_jk < 0.96:
        continue

    res_k, res_u = [rk[0]], [np.sqrt(rCk[0, 0])]
    final_jk = None

    for i in range(2, 101):
        try:
            inv_rCd = np.linalg.inv(rCd[1:i, 1:i] + 1e-12 * np.eye(i-1))
            jk = np.sqrt(Rd[1:i, 0] @ np.linalg.inv(Rd[1:i, 1:i]) @ Rd[1:i, 0])
        except np.linalg.LinAlgError:
            break

        if jk >= 0.95:
            if 0.94 <= jk <= 0.96:
                matching_seeds_only.append(seed)
            break

matching_seeds_only
```

```python
total_indices = list(range(620))

plt.figure(figsize = (12,5))
seed_list = [5, 369]
data = []
for j in range(2):
    random.seed(seed_list[j])
    
    selected_indices = random.sample(total_indices, 100)

    rk = k[selected_indices]
    rm = m[selected_indices]
    rCk = Ck[np.ix_(selected_indices, selected_indices)]
    rCm = Cm[np.ix_(selected_indices, selected_indices)]
    rCd = rCk + rCm
    rNames = [names[i] for i in selected_indices]

    std_dev = np.diag(np.sqrt(np.diag(rCd)))
    Rd = np.linalg.inv(std_dev)@(rCd)@np.linalg.inv(std_dev)

    row = Rd[0, :]
    closest_idx = np.argmin(np.abs(row - 0.7))
    remaining_indices = [i for i in range(100) if i not in (0, closest_idx)]
    random.shuffle(remaining_indices)
#     print(Rd[0,closest_idx])

    new_order = [0, closest_idx] + remaining_indices

    rk = rk[new_order]
    rm = rm[new_order]
    rCk = rCk[np.ix_(new_order, new_order)]
    rCm = rCm[np.ix_(new_order, new_order)]
    rCd = rCk + rCm
    rNames = [rNames[i] for i in new_order]

    Rd = np.linalg.inv(std_dev)@(rCd)@np.linalg.inv(std_dev)
    total_jk = np.sqrt(Rd[1:,0]@np.linalg.inv(Rd[1:,1:])@Rd[1:,0])

#     print(total_jk)

    res_k, res_u, jk_all, bname = [rk[0]], [np.sqrt(rCk[0,0])], [0.0], [rNames[0]]

    for i in range(2,101):
        pk = rk[0] - rCk[0,1:i]@np.linalg.inv(rCd[1:i,1:i])@(rk[1:i] - rm[1:i])
        res_k.append(pk)

        pCk = rCk[0,0] - rCk[0,1:i]@np.linalg.inv(rCd[1:i,1:i])@rCk[0,1:i]
        pu = np.sqrt(pCk)
        res_u.append(pu)

        jk = np.sqrt(Rd[1:i,0]@np.linalg.inv(Rd[1:i,1:i])@Rd[1:i,0])
        jk_all.append(jk)
        
        bname.append(rNames[i-1])
        
        if jk >= 0.95:
            break
    data.append([res_k, res_u, jk_all, bname])
#     print(jk)
#     print(len(res_k))
    plt.subplot(1,2,j+1)
    plt.errorbar(np.arange(len(res_k)), res_k, yerr=np.array(res_u)*1.96, fmt='o-', capsize=3)
    plt.axhline(rm[0], color='red', linestyle='--')
    plt.xlabel("Number of experiments used")
    plt.ylabel("keff with 95% conf. interv.")
    plt.legend(["msr keff", "keff est."])
    plt.grid()
    
plt.show()
```

```python
total_indices = list(range(620))

random.seed(181)

n = 70
selected_indices = random.sample(total_indices, n)

rk = k[selected_indices]
rm = m[selected_indices]
rCk = Ck[np.ix_(selected_indices, selected_indices)]
rCm = Cm[np.ix_(selected_indices, selected_indices)]

rCd = np.copy(rCk)
rCd[1:,1:] += rCm[1:,1:]

rNames = [names[i] for i in selected_indices]

std_dev = np.diag(np.sqrt(np.diag(rCd)))
Rd = np.linalg.inv(std_dev)@(rCd)@np.linalg.inv(std_dev)

std_dev = np.diag(np.sqrt(np.diag(rCk)))
Rk = np.linalg.inv(std_dev)@(rCk)@np.linalg.inv(std_dev)

res_k, res_u = [rk[0]], [np.sqrt(rCk[0,0])]

for i in range(2,n+1):
    pk = rk[0] - rCk[0,1:i]@np.linalg.inv(rCd[1:i,1:i])@(rk[1:i] - rm[1:i])
    res_k.append(pk)

    pCk = rCk[0,0] - rCk[0,1:i]@np.linalg.inv(rCd[1:i,1:i])@rCk[0,1:i]
    pu = np.sqrt(pCk)
    res_u.append(pu)
        
data = [res_k, res_u]
```

```python
ck_sort = np.argsort(Rk[:,0])[::-1]

jk_wo_msr = [0]
for j in range(n-1):
    njk = np.zeros((n,))
    for i in range(n):
        new_index = jk_wo_msr + [i]
        R = Rk[new_index][:,new_index]

        njk[i] = R[:j+1,-1]@np.linalg.inv(R[:j+1,:j+1] + np.eye(j+1)*1e-10)@R[-1,:j+1]
        
    arg_sort = np.argsort(njk)[::-1]
    jk_wo_msr.append(int(arg_sort[j+1]))
    
jk_w_msr = [0]
n = len(Rd)
for j in range(n-1):
    njk = np.zeros((n,))
    for i in range(n):
        new_index = jk_w_msr + [i]
        R = Rd[new_index][:,new_index]

        njk[i] = R[:j+1,-1]@np.linalg.inv(R[:j+1,:j+1])@R[-1,:j+1]

    arg_sort = np.argsort(njk)[::-1]
    jk_w_msr.append(int(arg_sort[j+1]))
```

```python
sortings = [ck_sort, jk_wo_msr, jk_w_msr]
for sorting in sortings:
    nk = rk[sorting]
    nm = rm[sorting]
    nCk = rCk[sorting][:,sorting]
    nCm = rCm[sorting][:,sorting]

    nCd = np.copy(nCk)
    nCd[1:,1:] += nCm[1:,1:]

    res_k, res_u = [nk[0]], [np.sqrt(nCk[0,0])]

    for i in range(2,n+1):
        pk = nk[0] - nCk[0,1:i]@np.linalg.inv(nCd[1:i,1:i] + np.eye(i-1)*1e-8)@(nk[1:i] - nm[1:i])
        res_k.append(pk)

        pCk = nCk[0,0] - nCk[0,1:i]@np.linalg.inv(nCd[1:i,1:i] + np.eye(i-1)*1e-8)@nCk[0,1:i]
        pu = np.sqrt(pCk)
        res_u.append(pu)

    data.append(res_k)
    data.append(res_u)
```
